```python
def parse_seeds(input):
    return list(map(lambda x: int(x), input.split(':')[1].split()))

# fpr part 2
def parse_seeds_2(input):
    output = list(map(lambda x: int(x), input.split(':')[1].split()))
    t_seeds = []
    for i in range(0, len(output), 2):
        t_seeds.append((output[i], output[i] + output[i + 1]))
    return t_seeds

def parse_map_categories(input):
    output = input.split()[0].split('-')
    return output[0], output[2]

def parse_map_values(input):
    output = list(map(lambda x: int(x), input.split()))
    return {'dest-start' : output[0],
            'dest-end' : output[0] + output[2],
            'src-start' : output[1],
            'src-end' : output[1] + output[2],
            'range-length' : output[2]}

def parse_map(input):
    output = input.split('\n')
    categories = parse_map_categories(output[0])
    values = list(map(lambda x: parse_map_values(x), output[1:]))
    return {'src' : categories[0],
            'dest': categories[1],
            'values' : values}

def parse_input(input):
    output = input.split('\n\n')
    maps = list(map(lambda x: parse_map(x), output[1:]))
    return parse_seeds(output[0]), parse_seeds_2(output[0]), maps

seeds, seeds2, maps = parse_input(input)

print(seeds2)
for m in maps:
    print(m)
```

```python
def get_map(src):
    return list(filter(lambda x : x['src'] == src, maps))[0]

#print(get_map('seed'))

def get_value(map, input_value):
    index = -1
    for i, m in enumerate(map['values']):
        if input_value in range(m['src-start'],m['src-end'],1):
            delta = input_value - m['src-start']
            return m['dest-start'] + delta
    return input_value

t_map = get_map('seed')
for i in range(46, 100, 1):
    print(i, get_value(get_map('seed'), i))
```

```python
def get_location_from_seed(seed):
    t_cat = 'seed'
    value = seed
    while t_cat != 'location':
        t_map = get_map(t_cat)
        value = get_value(t_map, value)
        t_cat = t_map['dest']
    return value

#get_location_from_seed(79)

for seed in seeds:
    print(get_location_from_seed(seed))
```

```python
def get_min_location():
    min = 99999999999
    for seed in seeds:
        t_loc = get_location_from_seed(seed)
        if t_loc < min:
            min = t_loc
    return min


get_min_location()
```
