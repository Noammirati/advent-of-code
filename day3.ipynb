```python
with open('day3.txt') as f:
    data = f.read().strip()

data = data.split('\n')
print(data)
```

```python
def get_adj(x,y):
    adj = []
    for i in range(max(0,x-1), min(len(data),x+2)):
        for ii in range(max(0, y-1), min(len(data[i]), y+2)):
            adj.append(data[i][ii])
    return adj

#get_adj(5,3)
get_adj(1,2)


def should_count(x,y):
    return len(list(filter(lambda x: not x.isdigit() and not x == '.', get_adj(x,y)))) != 0

print(True, should_count(0,2))
print(False, should_count(0,5))
```

```python
def get_numbers():
    numbers = {}
    for i in range(len(data)):
        prev_isNumber = False
        for ii in range(len(data[i])):
            if data[i][ii].isdigit():
                if not prev_isNumber:
                    numbers.update({(i,ii) : [data[i][ii], 1]})
                    prev_isNumber = True
                    prev_start = ii
                else:
                    prev = numbers[(i, prev_start)]
                    numbers.update({(i, prev_start) : [prev[0] + data[i][ii], prev[1] + 1]})
            elif prev_isNumber:
                prev_isNumber = False

    return numbers

get_numbers()
```

```python
def get_results():
    numbers = get_numbers()
    sum = 0
    for coord, value in numbers.items():
        count = False
        for ii in range(coord[1], coord[1] + value[1]):
            if should_count(coord[0], ii):
                count = True
        if count:
            sum += int(value[0])
    return sum

get_results()
```

```python
def get_adj_gears(x,y):
    return list(filter(lambda x: x == '*', get_adj(x,y)))

def get_adj_gears_2(x,y):
    adj = []
    for i in range(max(0,x-1), min(len(data),x+2)):
        for ii in range(max(0, y-1), min(len(data[i]), y+2)):
            if data[i][ii] == '*':
                adj.append((i, ii))
    return adj
```

```python
def set_gears():
    numbers = get_numbers()
    for coord, value in numbers.items():
        gears = {()}
        for ii in range(coord[1], coord[1] + value[1]):
            gears.update((get_adj_gears_2(coord[0], ii)))
        if len(gears) > 1:
            gears.remove(())
            value.append(gears)
    return numbers

set_gears()
```

```python
def get_all_gears(numbers_w_gears):
    gear_coords = {}
    for coord, value in numbers_w_gears.items():
        if len(value) == 3:
            for g in value[2]:
                if g in gear_coords.keys():
                    gear_coords[g].update({coord}) 
                else:
                    gear_coords.update({g: {coord}})
    return gear_coords

get_all_gears(set_gears())
```

```python
def get_result_2():
    numbers = set_gears()
    gears = get_all_gears(numbers)
    sum = 0
    for g, n in gears.items():
        print(g, n)
        if len(n) == 2:
            res = list(map(lambda x: numbers[x][0], n))
            print(res, int(res[0]) * int(res[1]))
            sum += int(res[0]) * int(res[1])

    return sum

get_result_2()
```
